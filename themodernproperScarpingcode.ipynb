```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os  
```

```python

def get_recipe(results):
    for job_element in results:
        recipe_name_list = job_element.find_all("h3", class_="posts-results__item--title")
        for recipe_name in recipe_name_list:
            recipe_list.append(recipe_name.text)
```

```python
recipe_list = []
for page in range(1,59):
    URL = "https://themodernproper.com/recipes/p"+str(page)
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find("div", class_="posts-results")
    get_recipe(results)
```

```python
def get_recipe(soup):    
    recipe_names = ""
    descriptions = ""
    prep_times = ""
    cook_times = ""
    serving_sizes = ""
    groups = ""
    calories = ""
    fats = ""
    cholesterol = ""
    sodium = ""
    carbohydrates = ""
    proteins = ""
    fibers = ""
    sugars = ""
    ingredients_list = ""
    instructions_list = ""
    
    results = soup.find_all("main", id = "main-content")
    for job_element in results:

        title = job_element.find("h1", class_="post-hero__title")
        if title:
            recipe_names = title.text
            print(recipe_names)
        description = job_element.find("p", class_="post-hero__summary")    
        if description:
            descriptions = description.text

        categories = job_element.find("div", class_="post-hero__category-wrapper")
        if categories:
            category = categories.find_all("a")
            for group in category:
                if group:
                    groups =groups+" "+ group.text

        prep = job_element.find("ul", class_="recipe-intro__stats-list")
        #print(prep)
        if prep:
            if "Serves:" in prep.text:
                servers = (re.search(r"Serves:\s*(\d+)", prep.text)).group(1)
                if servers:
                    serving_sizes = servers
            if "Prep Time:" in prep.text:
                prep_time = (re.search(r"Prep Time:\s*(\d+)", prep.text)).group(1)
                if prep_time:
                    prep_times = prep_time
            if "Cook Time:" in prep.text:
                cook_time = (re.search(r"Cook Time:\s*(\d+)", prep.text)).group(1)
                if cook_time:
                    cook_times = cook_time
        

        '''#ingredients = job_element.find("div", class_="recipe-ingredients")
        ingredients = job_element.find("ul", class_="recipe-ingredients__list")   
        print(ingredients)
        
        if ingredients:
            ingredients_list = ingredients.text'''
        for item in job_element.find_all('li', class_='recipe-ingredients__item'):
            amount = item.find('span', class_='recipe-ingredients__item--amount')
            ingredient = item.find('span', class_='recipe-ingredients__item--ingredient')
            if amount:
                ingredients_list += str(amount.text).strip()
            if ingredient:
                ingredients_list += (ingredient.text).strip()+"\n"
        #print(ingredients_list)

        instructions = job_element.find("div", itemprop = "recipeInstructions")
        if instructions:
            instructions_list = instructions.text

        nutritions = job_element.find("div", id = "nutrition")
        
        if nutritions:
            total_calories = (re.search(r"Calories\s*(\d+)", nutritions.text)).group(1)
            if total_calories:
                calories = total_calories
            protein = (re.search(r"Protein(\d+\s*g)", nutritions.text)).group(1)
            if protein:
                proteins = protein
            carb = (re.search(r"Carbohydrates(\d+\s*g)", nutritions.text)).group(1)
            if carb:
                carbohydrates = carb
            total_fat = (re.search(r"Total Fat(\d+\s*g)", nutritions.text)).group(1)
            if total_fat:
                fats = total_fat
            fiber = (re.search(r"Dietary Fiber(\d+\s*g)", nutritions.text)).group(1)
            if fiber:
                fibers = fiber
            chol = (re.search(r"Cholesterol(\d+\s*mg)", nutritions.text)).group(1)
            if chol:
                cholesterol = chol
            sod = (re.search(r"sodium(\d+\s*mg)", nutritions.text)).group(1)
            if sod:
                sodium =sod
            sugar = (re.search(r"Total Sugars(\d+\s*g)", nutritions.text)).group(1)
            if sugar:
                sugars = sugar

    data = {
    "Recipe Name": recipe_names,
    "Description": descriptions,
    "Prep Time": prep_times,
    "Cook Time": cook_times,
    "Serving Size": serving_sizes,
    "Group": groups,
    "Calories": calories,
    "Fat": fats,
    "Cholesterol": cholesterol,
    "Sodium": sodium,
    "Carbohydrate": carbohydrates,
    "Protein": proteins,
    "Sugar": sugars,
    "Fiber": fibers,
    "Ingredients": ingredients_list,
    "Instructions": instructions_list
    }
    return data
```

```python
#recipes__item__image

URL = "https://themodernproper.com/Beer-Battered-fish"

page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")


get_recipe(soup)
```

```python
df = pd.DataFrame(columns=["Recipe Name","Description","Prep Time","Cook Time","Serving Size","Group",
                           "Calories","Fat","Cholesterol","Sodium","Carbohydrate","Protein","Sugar","Fiber","Ingredients","Instructions"])

```

```python
for recipe_name in new_list:
    URL = "https://themodernproper.com/"+recipe_name
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    if soup:
        if "Page Not Found" not in (soup.find("title").text):
            df.loc[len(df)] = get_recipe(soup)
```
